```python
%reload_ext autoreload
%autoreload 2

from matplotlib import pyplot as plt
from utils import load_data, clean_data

titanic = load_data()
#titanic = clean_data(titanic)

titanic.head(10)

# Count passengers by class
class_counts = titanic["pclass"].value_counts().sort_index()

# Create bar plot
plt.figure(figsize=(6, 4))
class_counts.plot(kind="bar")

# Labels and title
plt.xlabel("Passenger Class")
plt.ylabel("Number of Male Passengers")
plt.title("Passenger Class Distribution (Men Only)")
plt.xticks(rotation=0)

# Show plot
plt.tight_layout()
plt.show()
titanic.head(10)
clean_data(titanic)
```

```python
import pandas as pd
ages = pd.to_numeric(titanic.get('age', pd.Series(dtype='float')), errors='coerce').dropna()

if ages.empty:
    print("No age data available to plot.")
else:
    plt.figure(figsize=(8, 5))
    plt.hist(ages, bins=30, edgecolor='k')
    plt.xlabel("Age")
    plt.ylabel("Count")
    plt.title("Titanic Passenger Age Distribution")
    plt.tight_layout()
    plt.show()
```
